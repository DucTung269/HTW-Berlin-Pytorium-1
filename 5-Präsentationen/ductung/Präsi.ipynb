```python
# year-to-year variability (percentage change) for each country
df_cleaned['2016-2017 Change (%)'] = ((df_cleaned['2017'] - df_cleaned['2016']) / df_cleaned['2016']) * 100
df_cleaned['2017-2018 Change (%)'] = ((df_cleaned['2018'] - df_cleaned['2017']) / df_cleaned['2017']) * 100
df_cleaned['2018-2019 Change (%)'] = ((df_cleaned['2019'] - df_cleaned['2018']) / df_cleaned['2018']) * 100
df_cleaned['2019-2020 Change (%)'] = ((df_cleaned['2020'] - df_cleaned['2019']) / df_cleaned['2019']) * 100

top_5_countries = df_cleaned.nlargest(5, '2020')

# Plot year-to-year variability for top producers
for _, row in top_5_countries.iterrows():
    variability = row[['2016-2017 Change (%)', '2017-2018 Change (%)', '2018-2019 Change (%)', '2019-2020 Change (%)']]
    plt.plot(variability.index, variability.values, marker='o', label=row['Country'])


plt.title('Year-to-Year Production Variability for Top 5 Producers (2016-2020)')
plt.xlabel('Year Range')
plt.ylabel('Change (%)')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```
